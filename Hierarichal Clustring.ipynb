# Hierarichal Clustring

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt  
from sklearn import datasets
```

```python
iris=datasets.load_iris()
```

```python
iris_data=pd.DataFrame(iris.data)
```

```python
iris_data.columns=iris.feature_names
```

```python
iris_data
```

```python
#standardization
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
```

```python
X_scaled = scaler.fit_transform(iris_data)
```

```python
X_scaled
```

```python
X_scaled.shape
```

```python
#appliying pca
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
pca
```

```python
pca_scaled=pca.fit_transform(X_scaled)
pca_scaled
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1])
```

```python
#Agglomerative clustring
import scipy.cluster.hierarchy as sc
#dindogram
plt.figure(figsize=(20,7))
plt.title("Dendogram")

#create dindogram
sc.dendrogram(sc.linkage(pca_scaled,method='ward')) ##eucliden distance
plt.title('dindogram')
plt.xlabel('Sample index')
plt.ylabel('Euciledean distance')
```

```python
from sklearn.cluster import AgglomerativeClustering
cluster=AgglomerativeClustering(n_clusters=2,linkage='ward')
cluster.fit(pca_scaled)
```

```python
cluster.labels_
```

```python
plt.scatter(pca_scaled[:,0],pca_scaled[:,1],c=cluster.labels_)
```

```python
from sklearn.metrics import silhouette_score
```

```python
silhouette_cofficients=[]
for k in range(2,11):
    agglo=AgglomerativeClustering(n_clusters=k,linkage='ward')
    agglo.fit(X_scaled)
    score=silhouette_score(X_scaled,agglo.labels_)
    silhouette_cofficients.append(score)
    
silhouette_cofficients
```

```python
plt.plot(range(2,11),silhouette_cofficients)
plt.xticks(range(2,11))
plt.xlabel("Number of clusters")
plt.ylabel("silhouette_cofficients")
plt.show()
```
