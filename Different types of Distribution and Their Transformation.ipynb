# Log Normal Distribution,Power Law Distribution

```python
mu, sigma = 3., 1. # mean and standard deviation
s = np.random.lognormal(mu, sigma, 1000)
```

```python
sns.histplot(s,kde=True)
```

```python
sns.histplot(np.log(s),kde=True)  ##converting it into normal distribution using loge
```

```python
## to check whether feature is guassian or normal distributed
## Q-Q plot
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab
def plot_data(sample):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
```

```python
## create a normal distributed data
s=np.random.normal(0.5,0.2,1000)
plot_data(s)
```

```python
plot_data(df['sepal_width'])  #also checking for sepal width feature
```

```python
##its clear ,it is a normal distribution
```

```python
mu, sigma = 3., 1. # mean and standard deviation
sample = np.random.lognormal(mu, sigma, 100)
```

```python
def plot_data_log(sample):   ##for checking log normal distribution
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample)
    plt.subplot(1,2,2)
    stat.probplot(np.log(sample),dist='norm',plot=pylab)
    plt.show()
```

```python
plot_data_log(sample)
```
